Fusiono las columnas del Dataframe general que considero necesarias para un EDA de las hojas hechos y victimas.

```python
import pandas as pd

# Asegúrate de que las rutas de archivo son correctas y accesibles
ruta_hechos_modificados = r'C:\Users\ASUS\PROYECTO INDIVIDUAL No 2 SINIESTROS VIALES\archivo_modificado.xlsx'
ruta_victimas = r'C:\Users\ASUS\PROYECTO INDIVIDUAL No 2 SINIESTROS VIALES\homicidios.xlsx'

# Cargar los DataFrames
df_hechos_modificado = pd.read_excel(ruta_hechos_modificados)
df_victimas = pd.read_excel(ruta_victimas, sheet_name='VICTIMAS')

# Realizar la fusión usando 'Id del hecho' como la clave
df_fusionado = pd.merge(df_hechos_modificado, df_victimas, left_on='Id del hecho', right_on='ID_hecho', how='left')

# Eliminar columnas innecesarias después de la fusión
df_fusionado.drop(columns=['ID_hecho', 'FECHA', 'AAAA', 'MM', 'DD', 'VICTIMA'], inplace=True)

# Renombrar las columnas restantes
df_fusionado.rename(columns={
    'ROL': 'Rol',
    'SEXO': 'Sexo',
    'EDAD': 'Edad',
    'FECHA_FALLECIMIENTO': 'Fecha_fallecimiento'
}, inplace=True)

# Asegurarse de que la columna 'Fecha_fallecimiento' es de tipo datetime
df_fusionado['Fecha_fallecimiento'] = pd.to_datetime(df_fusionado['Fecha_fallecimiento'], errors='coerce')

# Convertir la columna 'Fecha_fallecimiento' a formato de fecha (Año-Mes-Día), eliminando cualquier hora si es 00:00:00
df_fusionado['Fecha_fallecimiento'] = df_fusionado['Fecha_fallecimiento'].apply(lambda x: x.strftime('%Y-%m-%d') if not pd.isnull(x) else None)

# Mover la columna 'Acusado' después de 'Fecha_fallecimiento'
columnas = df_fusionado.columns.tolist()
indice_acusado = columnas.index('Acusado')
columnas.pop(indice_acusado)
indice_fecha_fallecimiento = columnas.index('Fecha_fallecimiento')
columnas.insert(indice_fecha_fallecimiento + 1, 'Acusado')
df_fusionado = df_fusionado.reindex(columns=columnas)

# Guardar el DataFrame fusionado en una nueva ubicación
ruta_guardado_fusionado = r'C:\Users\ASUS\PROYECTO INDIVIDUAL No 2 SINIESTROS VIALES\archivo_fusionado.xlsx'
df_fusionado.to_excel(ruta_guardado_fusionado, index=False)

print(f'DataFrame fusionado y guardado con éxito en: {ruta_guardado_fusionado}')
```

Me permito organizar la secuencia númerica de la columna Id del hecho.

```python
import pandas as pd

# Nueva ruta del archivo fusionado (Asegúrate de que la ruta sea correcta y el archivo exista)
ruta_archivo_fusionado = r'C:\Users\ASUS\PROYECTO INDIVIDUAL No 2 SINIESTROS VIALES\archivo_fusionado.xlsx'

try:
    # Cargar la hoja 'HECHOS' del DataFrame modificado
    df_hechos = pd.read_excel(ruta_archivo_fusionado, sheet_name='Sheet1')

    # Dividir la columna 'ID' en año y número de hecho. Asumiendo que 'Id del hecho' es similar a 'ID'
    df_hechos[['Año', 'Número']] = df_hechos['Id del hecho'].str.extract(r'(\d{4})-(\d+)')
    df_hechos['Año'] = df_hechos['Año'].astype(int)
    df_hechos['Número'] = df_hechos['Número'].astype(int)

    # Crear una nueva columna 'Id del hecho' con la secuencia reorganizada
    df_hechos['Id del hecho'] = df_hechos.groupby('Año').cumcount() + 1
    df_hechos['Id del hecho'] = df_hechos.apply(lambda x: f"{x['Año']}-{x['Id del hecho']:04d}", axis=1)

    # Cambiar 'SD' por '0' en columnas numéricas y 'Sin dato' en columnas de tipo string
    for columna in df_hechos.columns:
        if df_hechos[columna].dtype == 'object':
            df_hechos[columna] = df_hechos[columna].replace('SD', 'Sin dato')
        else:
            df_hechos[columna] = pd.to_numeric(df_hechos[columna], errors='coerce').fillna(0).astype(int)

    # Guardar el DataFrame modificado
    df_hechos.to_excel(ruta_archivo_fusionado, index=False)

    print("DataFrame modificado y guardado con éxito.")
except FileNotFoundError as e:
    print(f"Error al encontrar el archivo: {e}")
except Exception as e:
    print(f"Se produjo un error: {e}")
```

```python
import pandas as pd

# Nueva ruta del archivo fusionado (Asegúrate de que la ruta sea correcta y el archivo exista)
ruta_archivo_fusionado = r'C:\Users\ASUS\PROYECTO INDIVIDUAL No 2 SINIESTROS VIALES\archivo_fusionado.xlsx'

try:
    # Cargar la hoja 'HECHOS' del DataFrame modificado
    df_hechos = pd.read_excel(ruta_archivo_fusionado, sheet_name='Sheet1')

    # Dividir la columna 'ID' en año y número de hecho. Asumiendo que 'Id del hecho' es similar a 'ID'
    df_hechos[['Año', 'Número']] = df_hechos['Id del hecho'].str.extract(r'(\d{4})-(\d+)')
    df_hechos['Año'] = df_hechos['Año'].astype(int)

    # Crear una nueva columna 'Id del hecho' con la secuencia reorganizada
    df_hechos['Id del hecho'] = df_hechos.groupby('Año').cumcount() + 1
    df_hechos['Id del hecho'] = df_hechos.apply(lambda x: f"{x['Año']}-{x['Id del hecho']:04d}", axis=1)

    # Eliminar la columna 'Número'
    df_hechos.drop(columns=['Número'], inplace=True)

    # Cambiar 'SD' por '0' en columnas numéricas y 'Sin dato' en columnas de tipo string
    for columna in df_hechos.columns:
        if columna == 'Edad':
            df_hechos[columna] = df_hechos[columna].replace('Sin dato', 0)
        elif columna == 'Fecha_fallecimiento':
            df_hechos[columna].fillna('Desconocida', inplace=True)
        elif df_hechos[columna].dtype == 'object':
            df_hechos[columna] = df_hechos[columna].replace('SD', 'Sin dato')
        else:
            df_hechos[columna] = pd.to_numeric(df_hechos[columna], errors='coerce').fillna(0).astype(int)

    # Guardar el DataFrame modificado
    df_hechos.to_excel(ruta_archivo_fusionado, index=False)

    print("DataFrame modificado y guardado con éxito.")
except FileNotFoundError as e:
    print(f"Error al encontrar el archivo: {e}")
except Exception as e:
    print(f"Se produjo un error: {e}")
```

Total cantidad de hechos registrados por  año desde el 2016 hasta el 2021.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración para mejorar la estética de los gráficos
sns.set(style="whitegrid")

# Gráfico de barras para la cantidad de víctimas por año
plt.figure(figsize=(12, 7))
victimas_por_año = df_hechos.groupby('Año')['Numero de victimas'].sum().reset_index()
barplot = sns.barplot(x='Año', y='Numero de victimas', data=victimas_por_año, palette='viridis')
plt.title('Cantidad de Víctimas por Año')
plt.ylabel('Cantidad de Víctimas')
plt.xlabel('Año')

# Ajuste para anotar la cantidad de víctimas en cada barra
for index, row in victimas_por_año.iterrows():
    barplot.text(row.name, row['Numero de victimas'], round(row['Numero de victimas'], 2), color='black', ha="center")

plt.xticks(rotation=45)
plt.show()

# Gráfico de barras para la frecuencia de hechos por hora con anotaciones
plt.figure(figsize=(12, 7))
frecuencia_por_hora = df_hechos['Hora'].value_counts().sort_index()
sns.barplot(x=frecuencia_por_hora.index, y=frecuencia_por_hora.values, palette='viridis')
plt.title('Frecuencia de Hechos por Hora')
plt.xlabel('Hora del día')
plt.ylabel('Frecuencia')

# Añadir anotaciones en la parte superior de las barras
for index, value in enumerate(frecuencia_por_hora.values):
    plt.text(index, value, str(value), color='black', ha="center", va='bottom')

plt.xticks(range(24), [f'{h:02d}:00' for h in range(24)], rotation=45)
plt.tight_layout()
plt.show()

# Imprimir la hora más frecuente de accidentes y la cantidad de hechos
hora_mas_frecuente = df_hechos['Hora'].mode()[0]
hechos_por_hora = df_hechos['Hora'].value_counts()
print(f"La hora en que más frecuentemente ocurren hechos es: {hora_mas_frecuente}:00 hrs")
print(f"La cantidad de hechos a las {hora_mas_frecuente}:00 es de {hechos_por_hora[hora_mas_frecuente]} hechos.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Cargar la hoja 'HECHOS' del DataFrame
ruta_archivo = 'C:/Users/ASUS/PROYECTO INDIVIDUAL No 2 SINIESTROS VIALES/homicidios.xlsx'
df_hechos = pd.read_excel(ruta_archivo, sheet_name='HECHOS')

# Dividir la columna 'ID' en año y número de hecho y convertir a enteros
df_hechos[['Año', 'Número']] = df_hechos['ID'].str.split('-', expand=True)
df_hechos['Año'] = df_hechos['Año'].astype(int)
df_hechos['Número'] = df_hechos['Número'].astype(int)

# Ordenar el DataFrame por año y número de hecho
df_hechos.sort_values(by=['Año', 'Número'], inplace=True)

# Inicializar un diccionario para guardar los números faltantes por año
numeros_faltantes_por_año = {}

# Total global de números faltantes
total_numeros_faltantes = 0

# Iterar sobre cada año y detectar los números faltantes
for año in df_hechos['Año'].unique():
    numeros_del_año = df_hechos[df_hechos['Año'] == año]['Número']
    todos_los_numeros = set(range(1, numeros_del_año.max() + 1))
    numeros_faltantes = todos_los_numeros - set(numeros_del_año)
    cantidad_faltantes = len(numeros_faltantes)
    total_numeros_faltantes += cantidad_faltantes
    numeros_faltantes_por_año[año] = sorted(list(numeros_faltantes))

# Calcular el total de hechos por año teniendo en cuenta los números faltantes
total_hechos_por_año_corregido = {año: df_hechos[df_hechos['Año'] == año].shape[0] for año in df_hechos['Año'].unique()}

# Crear una gráfica de barras con el total de hechos por año corregido
plt.figure(figsize=(12, 6))
ax = plt.bar(total_hechos_por_año_corregido.keys(), total_hechos_por_año_corregido.values())
plt.title('Total de Hechos por Año (Corregido)')
plt.xlabel('Año')
plt.ylabel('Total de Hechos')
plt.xticks(rotation=45)

# Formatear el eje y para que solo muestre enteros
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.tight_layout()  # Ajusta automáticamente los parámetros de la subtrama
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ruta del archivo
ruta_archivo = 'C:/Users/ASUS/PROYECTO INDIVIDUAL No 2 SINIESTROS VIALES/homicidios.xlsx'

# Cargar los datos
df_hechos = pd.read_excel(ruta_archivo, sheet_name='HECHOS')

# Convertir 'FECHA' a datetime y extraer el año, filtrar por años 2016-2021
df_hechos['Año'] = pd.to_datetime(df_hechos['FECHA']).dt.year
df_hechos = df_hechos[df_hechos['Año'].between(2016, 2021)]

# Convertir 'N_VICTIMAS' a numérico y eliminar nulos
df_hechos['N_VICTIMAS'] = pd.to_numeric(df_hechos['N_VICTIMAS'], errors='coerce')
df_hechos.dropna(subset=['N_VICTIMAS'], inplace=True)

# Asegurar la correcta conversión de 'HORA' a formato de hora militar
if 'HH' in df_hechos.columns:
    df_hechos['HORA_CORREGIDA'] = pd.to_numeric(df_hechos['HH'], errors='coerce')
else:
    df_hechos['HORA_CORREGIDA'] = pd.to_datetime(df_hechos['HORA'], errors='coerce').dt.hour

# Eliminar filas con horas no válidas o nulas después de la conversión
df_hechos.dropna(subset=['HORA_CORREGIDA'], inplace=True)
df_hechos = df_hechos[df_hechos['HORA_CORREGIDA'].between(0, 23)]

# Configuración para mejorar la estética de los gráficos
sns.set(style="whitegrid")

# Gráfico de barras para la cantidad de víctimas por año
plt.figure(figsize=(12, 7))
victimas_por_año = df_hechos.groupby('Año')['N_VICTIMAS'].sum().reset_index()
barplot = sns.barplot(x='Año', y='N_VICTIMAS', data=victimas_por_año, palette='viridis')
plt.title('Cantidad de Víctimas por Año')
plt.ylabel('Cantidad de Víctimas')
plt.xlabel('Año')

# Anotar la cantidad de víctimas en cada barra
for index, row in victimas_por_año.iterrows():
    barplot.text(row.name, row.N_VICTIMAS, round(row.N_VICTIMAS, 2), color='black', ha="center")

plt.xticks(rotation=45)
plt.show()

# Gráfico de barras para la frecuencia de hechos por hora con anotaciones
plt.figure(figsize=(12, 7))
frecuencia_por_hora = df_hechos['HORA_CORREGIDA'].value_counts().sort_index()
sns.barplot(x=frecuencia_por_hora.index, y=frecuencia_por_hora.values, palette='viridis')
plt.title('Frecuencia de Hechos por Hora')
plt.xlabel('Hora del día')
plt.ylabel('Frecuencia')

# Añadir anotaciones en la parte superior de las barras
for index, value in zip(frecuencia_por_hora.index, frecuencia_por_hora.values):
    plt.text(index, value, str(value), color='black', ha="center", va='bottom')

plt.xticks(range(24), [f'{h:02d}:00' for h in range(24)], rotation=45)
plt.tight_layout()
plt.show()

# Imprimir la hora más frecuente de accidentes y la cantidad de hechos
hora_mas_frecuente = df_hechos['HORA_CORREGIDA'].mode()[0]
hechos_por_hora = df_hechos['HORA_CORREGIDA'].value_counts()
print(f"La hora en que más frecuentemente se accidentan las personas es: {hora_mas_frecuente}:00 hrs")
print(f"La hora con más hechos es a las {hora_mas_frecuente}:00 con {hechos_por_hora[hora_mas_frecuente]} hechos.")
```

```python
import pandas as pd

# Ruta del archivo
ruta_archivo = 'C:/Users/ASUS/PROYECTO INDIVIDUAL No 2 SINIESTROS VIALES/homicidios.xlsx'

# Cargar los datos
df_hechos = pd.read_excel(ruta_archivo, sheet_name='HECHOS')

# Asegurarse de que no haya valores nulos en 'FECHA' y 'HORA'
df_hechos.dropna(subset=['FECHA', 'HORA'], inplace=True)

# Intenta la conversión directa de 'HORA' si ya está en formato de hora, o ajusta según sea necesario
try:
    # Asumiendo que 'HORA' puede contener formatos de tiempo válidos como strings
    df_hechos['HORA'] = pd.to_datetime(df_hechos['HORA'], format='%H:%M:%S', errors='coerce').dt.time
except Exception as e:
    print(f"Error en la conversión de la hora: {e}")

# Combinar 'FECHA' y 'HORA' solo para filas sin errores de conversión
df_hechos['FECHA_HORA'] = pd.to_datetime(df_hechos['FECHA'].astype(str) + ' ' + df_hechos['HORA'].astype(str), errors='coerce')

# Eliminar cualquier fila que todavía tenga NaT en 'FECHA_HORA' después de la conversión
df_hechos.dropna(subset=['FECHA_HORA'], inplace=True)

# Ordenar por 'FECHA_HORA'
df_hechos.sort_values('FECHA_HORA', inplace=True)

# Calcular la diferencia de tiempo entre registros consecutivos.
df_hechos['DIFERENCIA_TIEMPO'] = df_hechos['FECHA_HORA'].diff()

# Convertir la diferencia de tiempo a horas.
df_hechos['DIFERENCIA_HORAS'] = df_hechos['DIFERENCIA_TIEMPO'].dt.total_seconds() / 3600

# Mostrar las diferencias de tiempo en horas.
print(df_hechos[['FECHA_HORA', 'DIFERENCIA_HORAS']].head())

# Crear un histograma de la diferencia de tiempo en horas.
plt.figure(figsize=(12, 7))
sns.histplot(df_hechos['DIFERENCIA_HORAS'].dropna(), kde=False, bins=24)
plt.title('Distribución de la Diferencia de Tiempo entre Accidentes')
plt.xlabel('Diferencia de Tiempo en Horas')
plt.ylabel('Frecuencia')
plt.show()

# Imprimir estadísticas descriptivas de la diferencia de tiempo.
print("Estadísticas de la diferencia de tiempo entre accidentes:")
print(df_hechos['DIFERENCIA_TIEMPO'].describe())
```
